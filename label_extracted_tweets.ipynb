```python
import nltk
```

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
a = 'I am very happpy that he got hurt'
```

```python
b = 'I am happy he is safe'
```

```python
sid.polarity_scores(a)
```

```python
sid.polarity_scores(b)
```

```python
dataset['scores'] = dataset['Text'].apply(lambda Text: sid.polarity_scores(Text))
```

```python
# the above line gives us the new column called as scores, we will only use the compoung score
```

```python
dataset['compound_score'] = dataset['scores'].apply(lambda d: d['compound'])
```

dataset['Text'].iloc[14]

```python
dataset['sentiment'] =  dataset['compound_score'].apply(lambda score: 1 if score >=0 else 0)
```
